```python
import heapq

class Graph:
    INF = float('inf')
    MAX = 100
    def __init__(self, MAX=100):
        self.V = 0 # 정점의 개수
        self.vertex = [""] * MAX 
        self.adj = [[self.INF] * MAX for _ in range(MAX)]#인접행렬
        self.MSTWeight = 0

    def prim(self, start):
        selected = [False] * self.V
        dist = [self.INF] * self.V
        dist[start] = 0

        for _ in range(self.V):
            u = self.GetMinVertex(selected, dist)
            selected[u] = True
            
            self.MSTWeight += dist[u]
            print(f"select = {self.vertex[u]}, W = {self.MSTWeight}")

            #dist[] 업데이트
            for v in range(self.V):
                if self.adj[u][v] != self.INF and not selected[v] and self.adj[u][v] < dist[v]:
                    dist[v] = self.adj[u][v]

    #MST에 포함되지 않은 정점 중에서 거리가 최소인 정점을 찾아서 리턴
    def GetMinVertex(self, selected, dist):
        min_dist = self.INF
        min_vertex = -1
        for v in range(self.V):
            if not selected[v] and dist[v] < min_dist:
                min_dist = dist[v]
                min_vertex = v
        return min_vertex

    def PrintGraph(self):
        print("vertex 수 =", self.V)
        for i in range(self.V):
            print(self.vertex[i], end="\t")
            for j in range(self.V):
                print(f"{self.adj[i][j]:8}", end=" ")
            print()

    def ReadGraph(self, filename):
        with open(filename, 'r') as file:
            lines = file.readlines()

        self.V = int(lines[0].strip())
        
        for i, line in enumerate(lines[1:]):
            elements = line.split()
            self.InsertVertex(i, elements[0])
            for j, weight in enumerate(elements[1:]):
                self.InsertEdge(i, j, int(weight))

    def InsertEdge(self, i, j, w):
        self.adj[i][j] = w

    def InsertVertex(self, index, name):
        self.vertex[index] = name


graph = Graph()
graph.ReadGraph("graph1.txt")
graph.PrintGraph()
graph.prim(0)  # 0번 노드부터 Prim 알고리즘 시작
```
