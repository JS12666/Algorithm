```python
import random
import time
import matplotlib.pyplot as plt

MAX = 1000000; #백만
N = 0;

#버블 정렬
def BubbleSort():
    for i in range(N - 1, 0, -1):
        for j in range(i):
            if(a[j] > a[j + 1]):
                t = a[j]
                a[j] = a[j+1]
                a[j + 1] = t

#합병정렬
def MergeSort(a, left, right):
    if left < right:
        mid = (left + right) // 2
        MergeSort(a, left, mid)
        MergeSort(a, mid + 1, right)
        Merge(a, left, mid, right)

def Merge(a, left, mid, right):
    i = left
    j = mid + 1
    k = left
    sorted_arr = [0] * (right - left + 1)
    
    while i <= mid and j <= right:
        if a[i] <= a[j]:
            sorted_arr[k - left] = a[i]
            i += 1
        else:
            sorted_arr[k - left] = a[j]
            j += 1
        k += 1

    while i <= mid:
        sorted_arr[k - left] = a[i]
        i += 1
        k += 1

    while j <= right:
        sorted_arr[k - left] = a[j]
        j += 1
        k += 1

    for i in range(left, right + 1):
        a[i] = sorted_arr[i - left]

def SWAP(a, i, j):
    tmp = a[i]
    a[i] = a[j]
    a[j] = tmp

def Partition(a, left, right):
    pivot = a[left]
    low = left + 1
    high = right
    
    while True:
        while low <= high and a[low] <= pivot:
            low += 1
        while low <= high and a[high] >= pivot:
            high -= 1
        if low < high:
            SWAP(a, low, high)
        else:
            break
    
    SWAP(a, left, high)
    return high

def QuickSort(a, left, right):
    if left < right:
        q = Partition(a, left, right)
        QuickSort(a, left, q - 1)
        QuickSort(a, q + 1, right)

def Graph(a):
    x = range(len(a))
    plt.bar(x, a, color='skyblue')
    plt.show()
```

```python
#데이터 개수
N = int(input("숫자 생성 갯수 : "))
a = [0] * N

for i in range(N):
    a[i] = random.randint(1, 5 * N - 1)

Graph(a)
```

```python
#시간 측정
#합병 정렬
start_time = time.perf_counter()
BubbleSort()
end_time = time.perf_counter()
#시간 계산
elapsed_time = end_time - start_time
elasped_ticks = int(elapsed_time * 1e6)

result = f"{elasped_ticks} Ticks, {elapsed_time * 1000} ms"
print("버블정렬 : " + result)
Graph(a)
```
